```python
from sklearn.model_selection import train_test_split

features = iris.data[:, 2:]
labels = iris.target
```

```python
# train_test_split 함수는 라벨링 데이터와 타겟 데이터를 사용자가 설정한 비율로 쪼개서 반환해주는 함수이다. stratify option을 사용할 경우 모집단의 비율을 유지한다.
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, stratify=labels, random_state=13)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
import numpy as np

# y_test 배열 내 값들의 중복을 제거한 후 각 원소 별 개수를 카운트합니다.
np.unique(y_test, return_counts=True)
```

```python
from sklearn.tree import DecisionTreeClassifier

iris_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
iris_tree.fit(X_train, y_train)

plt.figure(figsize=(12,8))
plot_tree(iris_tree)
```

```python
# train data를 학습한 후에 전체 데이터를 예측하는 것이 95% 정도 나왔다.
y_pred_tr = iris_tree.predict(iris.data[:, 2:])
accuracy_score(iris.target, y_pred_tr)
```

```python
from mlxtend.plotting import plot_decision_regions

plt.figure(figsize=(10, 5))

# 데이터를 나누고 max_depth를 줄이는 것이 과적합의 정도를 줄이는 데에 중요하구나
plot_decision_regions(X=X_train, y=y_train, clf=iris_tree, legend=2)
plt.show()
```

```python
# train data를 학습한 후에 테스트 데이터를 예측하는 정확도가 96%로 전체 데이터를 예측하는 95% 보다 높게 나왔다.
y_pred_test = iris_tree.predict(X_test)
accuracy_score(y_test, y_pred_test)
```

```python
# 사이즈, 라벨, 투명도
scatter_highlight_kwargs = {'s': 150, 'label': 'Test data', 'alpha': 0.9}
scatter_kwargs = {'s': 120, 'edgecolor': None, 'alpha': 0.9}

plt.figure(figsize=(10, 5))

# 어떤 테스트 데이터가 선택되었는지를 보여줍니다. features: iris.data[:, 2:], labels: iris.target,
# X_highlight: 표본 추출한 데이터, clf: 분류 모델(여기서는 max_depth를 2까지 주고 train_data를 학습시킨 분류기를 의미)
# max_depth를 얼만큼까지 주고 과적합을 줄이는 작업을 할 줄 알아야 한다.
plot_decision_regions(X=features, y=labels, X_highlight=X_test, clf=iris_tree, legend=2,
                    scatter_highlight_kwargs=scatter_highlight_kwargs,
                    scatter_kwargs=scatter_kwargs,
                    contourf_kwargs={'alpha':0.2}
                    )
```
