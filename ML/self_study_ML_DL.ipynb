```python
import matplotlib.pyplot as plt
```

```python
# 도미 데이터
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
```

```python
plt.scatter(bream_length, bream_weight)
plt.xlabel('bream length')
plt.ylabel('bream weight')
plt.show()
```

```python
# 빙어 데이터
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight

fish_data = [[l, w] for l, w in zip(length, weight)]
fish_target = [1]*len(bream_length) + [0]*len(smelt_weight)
```

```python
# K최근접 이웃: 해당 샘플 주위에 있는 5개(기본)를 선택해서 정답을 결정합니다. 빙어 데이터가 3개이고 도미 데이터가 2개이면 도미 데이터를 반환합니다.
from sklearn.neighbors import KNeighborsClassifier

# kn을 머신러닝 모델이라고 합니다. 알고리즘 자체를 모델이라고 하기도 합니다.
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)
kn.score(fish_data, fish_target)
```

```python
# numpy array로 반환
kn.predict([[30, 600]])
```

```python
# 주변의 49개의 데이터를 참고해라는 옵션입니다. 위에서 말한 기본값 5라는 것이 해당 옵션을 의미합니다.
kn49 = KNeighborsClassifier(n_neighbors=49)
kn49.fit(fish_data, fish_target)

# True
kn49.score(fish_data, fish_target) == (35/49)
```

```python
train_input = fish_data[:35]
train_target = fish_target[:35]

test_input = fish_data[35:]
test_target = fish_target[35:]
```

```python
kn = KNeighborsClassifier()
kn.fit(train_input, train_target)
kn.score(test_input, test_target)
```

```python
import numpy as np
input_arr = np.array(fish_data)
target_arr = np.array(fish_target)
```

```python
# 데이터 섞고 훈련 데이터셋과 테스트 데이터 셋 나누기
index = np.arange(49)
np.random.shuffle(index)

# 배열 슬라이싱: 인덱스가 담긴 nd.array를 제공해서 해당 인덱스의 값을 추출하는 방법
train_input = input_arr[index[:35]]
train_target = target_arr[index[:35]]

test_input = input_arr[index[35:]]
test_target = target_arr[index[35:]]

mixed_kn = KNeighborsClassifier()

plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(test_input[:,0], test_input[:,1])

plt.xlabel('length')
plt.ylabel('width')

mixed_kn.fit(train_input, train_target)
mixed_kn.score(test_input, test_target)
```
