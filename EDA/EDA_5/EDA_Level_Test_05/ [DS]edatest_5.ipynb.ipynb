```python
from datetime import datetime

def return_float_time(s1, s2):
    time1 = datetime.strptime(s1, "%H:%M")
    time2 = datetime.strptime(s2, "%H:%M")    
    time_difference_minutes = (time2 - time1).total_seconds() / 60    
    hours = int(time_difference_minutes // 60)
    minutes = int(time_difference_minutes % 60)    
    time_difference_display = hours + (minutes / 60)
    r = round(time_difference_display, 3)
    if r > 23:
        return 24
    return r
```
